```python
def load_results(path):
    original_res_df = pd.read_csv(path, sep="\t")
    res_df = original_res_df[original_res_df['metric'] == 'Dice']
    res_df = original_res_df[original_res_df['split'] == 'val'] # train val
    results = res_df.groupby('seg_mask')['value'].mean()
    #results["mDice"] = results.mean()
    results.index = [(i.replace('mask_', '')) if i.startswith('mask_') else i for i in results.index]
    #results = results[["mDice", *sorted([col for col in results.index if col != "mDice"])]]
    results[results == 0] = np.nan
    return results
```

```python
_dict = {
    "UNet": load_results(unet2d),
    "standard NCA": load_results(standard_nca),
 "sigmoid_nonorm_nca": load_results(sigmoid_nca),
 "steps10_nonorm_nca": load_results(steps10_nonorm_nca),
 "sigmoid_steps10_nonorm_nca": load_results(sigmoid_steps10_nonorm_nca),
 "fr1_sigmoid_steps10_nonorm_nca": load_results(fr1_sigmoid_steps10_nonorm_nca),
 "h128_sigmoid_steps10_nonorm_nca": load_results(h128_sigmoid_steps10_nonorm_nca),
 "large_nca": load_results(large_nca),
}
```

```python
df = pd.DataFrame(_dict)
df_sorted = df.sort_index(key=lambda x: x.astype(int))
df_sorted.loc['meanDice'] = df_sorted.mean()
df_sorted
```

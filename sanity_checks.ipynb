```python
import torch
from nnunet_ext.network_architecture.nca.OctreeNCA2D import OctreeNCA2D
from nnunet_ext.network_architecture.nca.LargeOctreeNCA2D import LargeOctreeNCA2D
torch.set_grad_enabled(False)
```

```python
octree = OctreeNCA2D(num_channels=8, num_input_channels=1, num_classes=1, hidden_size=5, 
                     num_steps=[ 8, 8], num_levels=2, pool_op_kernel_sizes=[[2,2], [2,2]], 
                     use_norm=False, fire_rate=0.5)
```

```python
x = torch.randn(1, 1, 128, 128)
```

```python
torch.random.manual_seed(42)
y = octree(x)
```

```python
large_octree = LargeOctreeNCA2D(octree, 3, 1, 4)
large_octree.return_all_logits = True
large_octree.do_ds = False
```

```python
torch.random.manual_seed(42)
y1 = large_octree(x)
```
